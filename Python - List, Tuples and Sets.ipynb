# Coding with List, Tuples and Sets

## Lists and Tuples

### Tuples

Tuples are comma separated elements in the parentheses of a command. 

Strings, integers, and floats can be found in tuples

Example: Tuple1 =('disco', 10, 1.2)

Tuples can be accessed with square brackets, within the brackets place teh corresponding number for the element in the tuple.

Example: Tuple1[1] would return 10

Tuples can be combined by adding them together

Example: Tuple2 = Tuple1 + ('Live', 10) would return ('disco', 10, 1.2,'Live', 10)

Tuples can be sliced (commands using :) and we can pull also use the length command on them (len()).


#### Tuples are immutable, they can't be changed once assigned. 

Tuples can contain other tuples and other complex data types. 

Example: NT = (1,2,('pop','rock'),(3,4),(1,2)))

How to access indexes within tuples that are within tuples with the following format

Variable[Main tuple index] [Mini tuple index]

To access 'pop' from the ('pop','rock') tuple in the previous example, We would use NT[2][0]



### List

List are mutable, they can be changed after being assigned.

Elements in list can also be accessed by using square brackets with the format 

Variable[List index] 

List can be combined together using the addition symbol (+)

List can also be sliced (:)

#### How list can be altered

Original List for Example: L=['Michael Jackson', 10.1, 1982]

Variable.extend() - Adds elements in the parentheses to the end of a list

L.extend(['pop',10])

L=['Michael Jackson', 10.1, 1982,'pop',10]

Variable.append() - Adds the list to the variable's list as one element

L.append(['pop',10])

L=['Michael Jackson', 10.1, 1982,['pop',10]]

List can be changed by assigning an index

L[0] = 'Chris Brown'

L= ['Chris Brown', 10.1, 1982]

del(Variable[]) - Deletes element from list

del(L[1])

L=['Michael Jackson', 1982]

Variable Element.split() - Seperates element in a variable to create a list

Name = ('Charity', 1989)

'Charity'.split(',') = 'C','h','a','r','i','t','y')
(This example was used with delimiter ',')

Aliasing - Assigning one list two names

## Sets

Sets are unordered, and they only have unique items meaning they will not contain duplicates when returned in output.

Sets use curly brackets {}

set() - Converts list to sets

Place list into set() to convert using the following format

set(List Variable)

### Set Operations

SetName.add - Adds items to sets

A = {'1','2','3'}

A.add('4')

Will return A = {'1','2','3','4'}

SetName.remove - Removes items from sets

A = {'1','2','3','4'}

A.remove('4')

Will return A = {'1','2','3'}

-> in <- command - Verifies if an item is in the set with format

Item in Set

if the item is in the set it will return True, if not False

-> & <- command - Finds the commonalities in two sets with format

Set & Set

SetName1.union(SetName2) - Joins 2 sets together

SetNameOfPossibleSubset.issubset(SetName1) - Checks to see if one set is a subset of another


### Dictionaries

Terms

Key - Index by label and are immutable

Element/Values - Definition or meaning

Dictionaries use the following format(s)

{'Key':Value}

Example: {'key1':1,'key2':'2','key3':[3,3,3],'key4':(4,4,4),('key5'):5, (0, 1): 6}

Keys and Pairs are separated by commas. 

#### Dictionary Operations

DICT[KEY] - Find value of one key in the DICT

DICT[VALUE] - Find key belonging to the value in the DICT

DICT[Key]=Variable - adds element to the dictionary list

del(DICT[Key]) - removes element from the dictionary list

DICT.keys() - List all keys in dictionary

DICT.values() - List all values in dictionary

-> in <- command finds out if key or value is in the DICT with format
